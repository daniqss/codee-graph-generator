```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


def bar_plot(df, col, title='Bar plot', include_total=True, max_range=0):
    df = df.sort_values(by=['Checker', col], ascending=[True, True])

    fig, ax = plt.subplots(figsize=(12, 6))

    ax.bar(df['Checker'], df[col], color=plt.cm.Paired.colors)
    ax.set_title(f"{title} by checker and {col}", pad=25)
    ax.set_xticklabels(df['Checker'], rotation=75, fontsize=8)
    return fig
```

```python
bar_plot(d3, 'Priority')
bar_plot(d3, '#')
```
